## Map, Filter, Reduce

```python
# map() — apply function to every item

numbers = [1, 2, 3, 4, 5]

# Normal way
squares = []
for n in numbers:
    squares.append(n ** 2)

# map() way — cleaner
squares = list(map(lambda x: x ** 2, numbers))
print(squares)   # [1, 4, 9, 16, 25]
```

```python
# filter() — keep items that match condition

numbers = [1, 2, 3, 4, 5, 6, 7, 8]

evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)   # [2, 4, 6, 8]
```

```python
# reduce() — reduce list to single value
from functools import reduce

numbers = [1, 2, 3, 4, 5]

total = reduce(lambda x, y: x + y, numbers)
print(total)   # 15  (1+2+3+4+5)

product = reduce(lambda x, y: x * y, numbers)
print(product)  # 120  (1*2*3*4*5)
```

```python
# All 3 together — real example

from functools import reduce

marks = [92, 45, 78, 55, 88, 39, 95]

# Step 1 — filter passing marks only (>=60)
passing = list(filter(lambda x: x >= 60, marks))
print(passing)    # [92, 78, 88, 95]

# Step 2 — add 5 bonus marks to each
boosted = list(map(lambda x: x + 5, passing))
print(boosted)    # [97, 83, 93, 100]

# Step 3 — get total of boosted marks
total = reduce(lambda x, y: x + y, boosted)
print(total)      # 373
```

```python
# Task 1: Given prices = [120, 450, 89, 300, 670, 55, 200] —
# use filter to keep prices above 150, then use map to apply 10% discount on them.

import math

prices = [120, 450, 89, 300, 670, 55, 200]

above150 = list(filter(lambda x: x > 150, prices))

print(above150)

disc = list(map(lambda x: x * 0.9, above150))
print(disc)
```

```python
# Task 2: Given names = ["rashid", "ali", "sara", "ahmed"] —
# use map to capitalize all names.

names = ["rashid", "ali", "sara", "ahmed"]

capt = list(map(lambda x: x.upper(), names))

print(capt)
```
