 "What is Abstraction?"

"Abstraction means hiding complex implementation details and showing only the necessary interface. In Python we implement abstraction using the ABC module. An abstract class cannot be instantiated directly. Any child class that inherits from it must implement all abstract methods, otherwise Python throws an error."

```python
# Abstraction means — hiding complex implementation and showing only what is necessary. Real life example — you drive a car. You use steering, accelerator, brake.
# You don't know how engine works internally. That complexity is hidden. That is abstraction.
```

```python
# How to implement in Python — ABC module

from abc import ABC, abstractmethod

class Shape(ABC):           # abstract class
    @abstractmethod
    def area(self):         # abstract method — no implementation
        pass

    @abstractmethod
    def perimeter(self):    # abstract method
        pass
```

```python
# Abstract class — cannot create object directly.
# Abstract method — must be overridden in child class.
pythons = Shape()   # ❌ ERROR — cannot instantiate abstract class
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):                           # must implement
        print(f"Area: {self.length * self.width}")

    def perimeter(self):                      # must implement
        print(f"Perimeter: {2 * (self.length + self.width)}")

class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        print(f"Area: {3.14 * self.r * self.r}")

    def perimeter(self):
        print(f"Perimeter: {2 * 3.14 * self.r}")

r = Rectangle(5, 3)
r.area()        # Area: 15
r.perimeter()   # Perimeter: 16

c = Circle(7)
c.area()        # Area: 153.86
c.perimeter()   # Perimeter: 43.96
```

```python
from abc import ABC, abstractmethod

class Payment(ABC):

  @abstractmethod
  def pay(self, amount):
    pass

  @abstractmethod
  def get_payment_details(self):
    pass

class CreditCard(Payment):
  def __init__(self, card_number, owner):
    self.card_number = card_number
    self.owner = owner

  def pay(self, amount):
      print(f"{self.owner} paid {amount} using Credit Card {self.card_number}")

  def get_payment_details(self):
    print(f"Owner: {self.owner}, Card: {self.card_number}")

class UPI(Payment):
  def __init__(self, upi_id, owner):
    self.upi_id = upi_id
    self.owner = owner

  def pay(self, amount):
    print(f"{self.owner} paid {amount} using UPI  {self.upi_id}")


  def get_payment_details(self):
    print(f"Owner: {self.owner}, UPI: {self.upi_id}")

class Cash(Payment):
  def __init__(self, owner):
    self.owner = owner

  def pay(self, amount):
    print(f"{self.owner} paid {amount} using Cash")


  def get_payment_details(self):
    print(f"Owner: {self.owner}")


obj = CreditCard(9727, "Rashid")
obj.pay(10000)
obj.get_payment_details()

obj = UPI(9721, "Rashid")
obj.pay(10000)
obj.get_payment_details()

obj = Cash("Rashid")
obj.pay(10000)
obj.get_payment_details()
```

Encapsulation vs Abstraction:

"Encapsulation hides data using private variables and protects it with getters and setters. Abstraction hides complexity and shows only necessary functionality like how we drive a car without knowing the engine."

Inheritance vs Polymorphism:

"Inheritance allows a child class to acquire properties and methods of a parent class. Polymorphism means the same method behaves differently depending on which class calls it."
