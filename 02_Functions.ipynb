## 📌 What is a Function?

A function is a **reusable block of code** that performs a specific task.  
Instead of writing the same code again and again, we define it once with `def` and call it whenever needed.  
Functions can take inputs called **parameters** and return **outputs**.

## 1️⃣ Basic Functions

**Syntax:**
```python
def function_name(parameters):
    # code block
    return value
```

````python
# Basic function — takes a name and returns a greeting
def greet(name):
    return f"Hello, {name}!"

print(greet("Rashid"))   # Output: Hello, Rashid!
print(greet("Ali"))      # Output: Hello, Ali!
````

---
## 2️⃣ Default Arguments

Default arguments are used when the caller **does not provide a value**.  
They must always come **after** required arguments.

````python
# Default argument — role has a default value of 'Student'
def greet(name, role="Student"):
    return f"Hello {name}, you are a {role}"

print(greet("Rashid"))                 # uses default role
print(greet("Rashid", "ML Engineer"))  # overrides default
````

---
## 3️⃣ `*args` — Variable Positional Arguments

`*args` is used when you **don't know how many arguments** will be passed.  
All values are collected into a **tuple**.

````python
# *args — accepts any number of positional arguments
def add_all(*numbers):
    return sum(numbers)

print(add_all(1, 2, 3))           # 6
print(add_all(1, 2, 3, 4, 5))    # 15
print(add_all(10, 20))            # 30
````

````python
# Practical use of *args — calculate average marks
def average_marks(*marks):
    avg = sum(marks) / len(marks)
    return avg

print(average_marks(92, 83, 90, 79, 76, 82))  # average of 6 marks
````

---
## 4️⃣ `**kwargs` — Variable Keyword Arguments

`**kwargs` is used when you want to accept **named arguments** of any number.  
All values are collected into a **dictionary**.

````python
# **kwargs — accepts any number of keyword arguments as a dictionary
def show_details(**info):
    for key, value in info.items():
        print(f"{key} : {value}")

show_details(name="Rashid", age=21, goal="ML Engineer")
````

## 5️⃣ Lambda Functions

Lambda is a **small, anonymous, one-line function**.  
It is commonly used with `map()`, `filter()`, and `sorted()`.

**Syntax:** `lambda parameters: expression`

````python
# Lambda — one-line function
square = lambda x: x ** 2
print(square(5))   # 25
print(square(9))   # 81
````

````python
# Lambda with map() — apply a function to every item in a list
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x ** 2, numbers))
print(squared)  # [1, 4, 9, 16, 25]

# Lambda with filter() — keep only items that match condition
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)    # [2, 4]
````

## 6️⃣ Scope — Local, Global, Nonlocal

**Scope** defines where a variable can be accessed in the code.

| Scope | Where defined | Where accessible |
|-------|--------------|------------------|
| Local | Inside a function | Only inside that function |
| Global | Outside all functions | Everywhere |
| Nonlocal | Outer function | Inner (nested) function |

````python
# Local scope — variable only exists inside the function
def my_function():
    x = 10          # local variable
    print(x)        # works fine inside

my_function()
# print(x)         # ERROR — x does not exist outside
````

````python
# Global scope — variable accessible everywhere
name = "Rashid"    # global variable

def print_name():
    print(name)    # can READ global variable

print_name()       # Rashid
````

````python
# global keyword — needed to MODIFY a global variable inside a function
count = 0

def increment():
    global count   # tell Python to use the global variable
    count += 1

increment()
increment()
increment()
print(count)       # 3
````

````python
# nonlocal — used inside nested functions to access outer function's variable
def outer():
    message = "Hello"

    def inner():
        nonlocal message       # access outer function's variable
        message += " Rashid"  # modify it
        print(message)

    inner()

outer()    # Hello Rashid
````

## 7️⃣ Return Values — Multiple & Dictionary

A function can return **multiple values** as a tuple, or return a **dictionary** for structured data.

````python
# Returning multiple values — unpacked as tuple
def analyze_marks(marks):
    total = sum(marks)
    average = total / len(marks)
    status = "Passed" if average >= 60 else "Failed"
    return total, average, status    # returns 3 values

total, average, status = analyze_marks([69, 81, 78, 54])
print(f"Total   : {total}")
print(f"Average : {average}")
print(f"Status  : {status}")
````

````python
# Returning a dictionary — common in real projects
def get_bmi_report(name, height_cm, weight):
    height_m = height_cm / 100              # convert cm to meters
    bmi = weight / (height_m ** 2)          # BMI formula

    if bmi < 18.5:
        category = "Underweight"
    elif bmi < 25:
        category = "Normal"
    else:
        category = "Overweight"

    return {
        "name"     : name,
        "bmi"      : round(bmi, 2),
        "category" : category
    }

result = get_bmi_report("Rashid", 174, 80)

print(f"Name     : {result['name']}")
print(f"BMI      : {result['bmi']}")
print(f"Category : {result['category']}")
````

## 8️⃣ Practical Projects

These are real projects built using all the function concepts above.

````python
# Project 1 — BMI Calculator with user input
def bmi_calculator(height_cm, weight):
    height_m = height_cm / 100
    bmi = weight / (height_m ** 2)

    if bmi < 18.5:
        category = "Underweight"
    elif bmi < 25:
        category = "Normal"
    else:
        category = "Overweight"

    return round(bmi, 2), category

# Test with different values
print(bmi_calculator(175, 70))   # Normal
print(bmi_calculator(175, 45))   # Underweight
print(bmi_calculator(175, 90))   # Overweight
````

````python
# Project 2 — Student Result System using *args
def student_result(name, *marks):
    total   = sum(marks)
    average = total / len(marks)
    status  = "Passed" if average >= 60 else "Failed"

    return {
        "name"    : name,
        "total"   : total,
        "average" : round(average, 2),
        "status"  : status
    }

report = student_result("Rashid", 92, 83, 90, 79, 76, 82)

for key, value in report.items():
    print(f"{key:10} : {value}")
````
