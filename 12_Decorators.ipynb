## Decorators


A decorator is a function that wraps another function to add extra behavior without changing the original function.
Real life example — a security guard at a building. The building is your function. The guard adds extra behavior (checking ID) before letting anyone in. The building itself doesn't change.

"A decorator is a function that wraps another function to add extra behavior without modifying the original function. We apply decorators using the @ symbol. Decorators are commonly used for logging, authentication, timing, and caching. Inside the decorator we define a wrapper function that runs code before and after the original function."

```python
def my_decorator(func):
    def wrapper():
        print("Before function runs")
        func()                        # call original function
        print("After function runs")
    return wrapper

def greet():
    print("Hello Rashid!")

# Apply decorator manually
greet = my_decorator(greet)
greet()

# Output:
# Before function runs
# Hello Rashid!
# After function runs
```

Using @ syntax — cleaner way


```python
def my_decorator(func):
    def wrapper():
        print("Before function runs")
        func()
        print("After function runs")
    return wrapper

@my_decorator          # same as greet = my_decorator(greet)
def greet():
    print("Hello Rashid!")

greet()
```

```python
# Decorator with arguments

def my_decorator(func):
    def wrapper(*args, **kwargs):    # accept any arguments
        print("Before function")
        result = func(*args, **kwargs)  # pass arguments to original
        print("After function")
        return result
    return wrapper

@my_decorator
def add(a, b):
    return a + b

print(add(5, 3))
```

```python
# Real world example — Timer decorator

import time

def timer(func):
    def wrapper(*args, **kwargs):
        start  = time.time()
        result = func(*args, **kwargs)
        end    = time.time()
        print(f"{func.__name__} took {round(end - start, 4)} seconds")
        return result
    return wrapper

@timer
def calculate_squares(n):
    return [i ** 2 for i in range(n)]

calculate_squares(21)
```

```python
# Real world example — Login check decorator

def login_required(func):
    def wrapper(user, *args, **kwargs):
        if not user.get("logged_in"):
            print("Access denied. Please login first.")
            return
        return func(user, *args, **kwargs)
    return wrapper

@login_required
def dashboard(user):
    print(f"Welcome to dashboard, {user['name']}!")

user1 = {"name": "Rashid", "logged_in": True}
user2 = {"name": "Ali",    "logged_in": False}

dashboard(user1)   # Welcome to dashboard, Rashid!
dashboard(user2)   # Access denied. Please login first.
```

```python
# Task 1: Write a decorator uppercase_result that converts the return value of any function to uppercase.
# Apply it to a function get_name() that returns "rashid".

def uppercase_result(func):
  def wrapper():
    result = func()
    return result.upper()
  return wrapper

@uppercase_result
def get_name():
  return "rashid"

print(get_name())
```

```python
# Task 2: Write a decorator validate_age that checks if age argument is between 0 and 120 before the function runs.
# If not — print "Invalid age" and stop. Apply it to a function register(age).

def validate_age(func):
  def wrapper(age):
    if age < 0 or age > 120:
      print("Invalid age")
      return
    return func(age)
  return wrapper

@validate_age
def register(age):
  print(age)

register(25)
register(210)
register(-25)
```
