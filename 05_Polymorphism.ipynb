What is Polymorphism?

Polymorphism means the same method name behaves differently based on the object calling it. In Python we achieve this through method overriding — where a child class redefines a parent class method. We also have duck typing where Python doesn't check the object type, it just checks if the method exists and calls it.

```python
# Poly = many. Morphism = forms. Same method name, different behavior.
```

Type 1 — Method Overriding
Child class redefines parent's method.

```python
class Animal:
    def sound(self):
        print("Some animal sound")

class Dog(Animal):
    def sound(self):          # overrides parent method
        print("Woof!")

class Cat(Animal):
    def sound(self):          # overrides parent method
        print("Meow!")

class Cow(Animal):
    def sound(self):          # overrides parent method
        print("Moo!")

# Polymorphism in action
animals = [Dog(), Cat(), Cow()]

for animal in animals:
    animal.sound()            # same method, different output
# Woof!
# Meow!
# Moo!
```

```python
class Car:
    def move(self):
        print("Car drives on road")

class Boat:
    def move(self):
        print("Boat sails on water")

class Plane:
    def move(self):
        print("Plane flies in sky")

def start_journey(vehicle):   # doesn't care what type
    vehicle.move()            # just calls move()

start_journey(Car())
start_journey(Boat())
start_journey(Plane())
```

```python
# Task 1: Create a parent class Shape with method area() that prints "Calculating area".
# Create 3 child classes — Circle, Rectangle, Triangle. Each overrides area() with correct formula.
# Create a list of all 3 objects and loop through calling area() on each.
# Formulas:

# Circle — 3.14 * r * r
# Rectangle — length * width
# Triangle — 0.5 * base * height

class Shape:
  def area(self):
    print("Calculating area")

class Circle(Shape):
  def __init__(self, r):
    self.r = r

  def area(self):
    print(f"The area of Circle : {3.14 * self.r * self.r}")

class Rectangle(Shape):
  def __init__(self, lenght, width):
    self.lenght = lenght
    self.width = width

  def area(self):
    print(f"The area of Ractangle : {self.lenght * self.width}")

class Triangle(Shape):
  def __init__(self, base, height):
    self.base = base
    self.height = height

  def area(self):
    print(f"The area of Triangle : {0.5 * self.base * self.height}")

Shapes = [Circle(7), Rectangle(8,8), Triangle(15,10)]

for shape in Shapes:
  shape.area()
```

Type 2 — Duck Typing
Python doesn't care about the class type. If the method exists — it runs.

```python
class Car:
    def move(self):
        print("Car drives on road")

class Boat:
    def move(self):
        print("Boat sails on water")

class Plane:
    def move(self):
        print("Plane flies in sky")

def start_journey(vehicle):   # doesn't care what type
    vehicle.move()            # just calls move()

start_journey(Car())
start_journey(Boat())
start_journey(Plane())
```
