## рџ“Њ What is a Loop?

Instead of writing the same code 100 times, we use a loop.

## 1пёЏвѓЈ `for` Loop

A `for` loop is used to iterate over a **sequence** (list, string, range, tuple, etc.).

**Syntax:**
```python
for variable in sequence:
    # code block
```

````python
# Basic for loop вЂ” iterate over a list
fruits = ["apple", "banana", "mango", "orange"]

for fruit in fruits:
    print(fruit)
````

````python
# for loop вЂ” iterate over a string
name = "Rashid"

for letter in name:
    print(letter)
````

## 2пёЏвѓЈ `range()` вЂ” Properly

`range()` generates a sequence of numbers.

| Syntax | Output |
|--------|--------|
| `range(5)` | 0, 1, 2, 3, 4 |
| `range(1, 6)` | 1, 2, 3, 4, 5 |
| `range(0, 10, 2)` | 0, 2, 4, 6, 8 вЂ” step of 2 |
| `range(10, 0, -1)` | 10, 9, 8 ... 1 вЂ” reverse |

````python
# range() examples
print("range(5):")
for i in range(5):
    print(i, end=" ")

print("\n\nrange(1, 6):")
for i in range(1, 6):
    print(i, end=" ")

print("\n\nrange(0, 10, 2) вЂ” step 2:")
for i in range(0, 10, 2):
    print(i, end=" ")

print("\n\nrange(10, 0, -1) вЂ” reverse:")
for i in range(10, 0, -1):
    print(i, end=" ")
````

## 3пёЏвѓЈ `while` Loop

A `while` loop runs **as long as the condition is True**.  
Use it when you **don't know** how many times the loop should run.

**Syntax:**
```python
while condition:
    # code block
```

````python
# Basic while loop
count = 1

while count <= 5:
    print(f"Count: {count}")
    count += 1    # IMPORTANT: always update condition variable to avoid infinite loop

print("Loop finished")
````

````python
# while True вЂ” runs forever until break is used
while True:
    number = int(input("Enter a number (negative to stop): "))
    if number < 0:
        print("Negative number entered. Stopping.")
        break
    print(f"You entered: {number} вЂ” keep going!")
````

## 4пёЏвѓЈ `break`, `continue`, `pass`

| Keyword | What it does |
|---------|-------------|
| `break` | Exits the loop completely |
| `continue` | Skips current iteration, goes to next |
| `pass` | Does nothing вЂ” placeholder |


````python
# break вЂ” exit loop completely
print("break example:")
for i in range(10):
    if i == 5:
        break           # stop when i reaches 5
    print(i, end=" ")  # prints 0 1 2 3 4
````

````python
# continue вЂ” skip current iteration
print("continue example вЂ” skip even numbers:")
for i in range(10):
    if i % 2 == 0:
        continue        # skip even numbers
    print(i, end=" ")  # prints only odd: 1 3 5 7 9
````

````python
# pass вЂ” placeholder, does nothing
for i in range(5):
    pass    # no error, just empty loop вЂ” used as placeholder while building code

print("Loop with pass completed")
````

## 5пёЏвѓЈ `else` with Loops

The `else` block runs **only when the loop completes normally** (without `break`).  
If `break` is used, the `else` block is **skipped**.

````python
# else runs when loop completes normally
for i in range(5):
    print(i, end=" ")
else:
    print("\nLoop completed normally!")   # this runs
````

````python
# else does NOT run if break is used
for i in range(5):
    if i == 3:
        break
    print(i, end=" ")
else:
    print("\nThis will NOT print")   # skipped because of break
````

## 6пёЏвѓЈ `enumerate()` вЂ” Loop with Index

`enumerate()` gives both the **index** and **value** while looping.  
No need to manually track index with a counter variable.

````python
# enumerate вЂ” get index and value together
foods = ["Biryani", "Chicken Chilli", "White Chicken", "Curry", "Naan"]

# Basic way
print("Basic enumerate:")
for index, name in enumerate(foods):
    print(f"{index} : {name}")

# Numbered list starting from 1
print("\nNumbered list:")
for index, name in enumerate(foods, start=1):
    print(f"{index}. {name}")
````

## 7пёЏвѓЈ `zip()` вЂ” Loop Two Lists Together

`zip()` combines two or more lists and loops through them **together**.  

вљ пёЏ **Important rule:** Use different names for list variables and loop variables.

````python
# zip вЂ” loop two lists at the same time
# NOTE: list variables are 'names' and 'marks'
#       loop variables are 'name' and 'mark' вЂ” different names to avoid overwriting

names = ["Rashid", "Rahul", "Rohit", "Vijay"]
marks = [80, 75, 57, 69]

for name, mark in zip(names, marks):
    if mark >= 60:
        print(f"{name} : Pass")
    else:
        print(f"{name} : Fail")
````

## 8пёЏвѓЈ List Comprehension

List comprehension is a **short, clean way** to create lists using a single line.  
Very commonly asked in interviews.

**Syntax:**
```python
new_list = [expression for item in iterable if condition]
```

````python
# Normal way vs List comprehension

# Normal way вЂ” 4 lines
squares_normal = []
for i in range(1, 6):
    squares_normal.append(i ** 2)
print("Normal way:", squares_normal)

# List comprehension вЂ” 1 line
squares_comp = [i ** 2 for i in range(1, 6)]
print("Comprehension:", squares_comp)
````

````python
# List comprehension with condition
evens = [i for i in range(20) if i % 2 == 0]
print("Even numbers:", evens)

# All numbers divisible by 3 between 1 and 50
div_by_3 = [x for x in range(1, 51) if x % 3 == 0]
print("Divisible by 3:", div_by_3)
````

## 9пёЏвѓЈ Dict Comprehension

Same idea as list comprehension but creates a **dictionary**.

**Syntax:**
```python
new_dict = {key: value for item in iterable}
```

````python
# Dict comprehension вЂ” create a dictionary of squares
squares_dict = {x: x ** 2 for x in range(1, 6)}
print(squares_dict)   # {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

# Dict comprehension from two lists using zip
names = ["Rashid", "Ali", "Sara"]
marks = [92, 78, 85]

result = {name: mark for name, mark in zip(names, marks)}
print(result)   # {'Rashid': 92, 'Ali': 78, 'Sara': 85}
````

## рџ”џ Practical Projects

````python
# Project 1 вЂ” Student Pass/Fail Report using zip and enumerate
names  = ["Rashid", "Rahul", "Rohit", "Vijay", "Sara"]
marks  = [80, 75, 57, 69, 91]

print("=" * 35)
print(f"{'No.':<5} {'Name':<10} {'Marks':<8} {'Status'}")
print("=" * 35)

for index, (name, mark) in enumerate(zip(names, marks), start=1):
    status = "Pass" if mark >= 60 else "Fail"
    print(f"{index:<5} {name:<10} {mark:<8} {status}")

print("=" * 35)
````

````python
# Project 2 вЂ” Number guessing game using while loop
import random

secret  = random.randint(1, 10)   # random number between 1 and 10
attempts = 0

print("Guess the number between 1 and 10!")

while True:
    guess    = int(input("Your guess: "))
    attempts += 1

    if guess < secret:
        print("Too low! Try again.")
    elif guess > secret:
        print("Too high! Try again.")
    else:
        print(f"Correct! You guessed it in {attempts} attempts.")
        break
````

````python
# Project 3 вЂ” Shopping cart using list comprehension
prices = [120, 450, 89, 300, 670, 55, 200]

# Items above 200 rupees
expensive = [p for p in prices if p > 200]
print("Expensive items:", expensive)

# Apply 10% discount to all items above 200
discounted = [round(p * 0.9, 2) if p > 200 else p for p in prices]
print("After discount:", discounted)

print(f"Original total  : {sum(prices)}")
print(f"Discounted total: {sum(discounted)}")
````
