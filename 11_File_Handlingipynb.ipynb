## File Handling

"File handling in Python is done using the open() function with different modes — r for read, w for write, a for append. The with statement is the correct way to open files because it automatically closes the file after the block completes, even if an error occurs. For CSV files we use the csv module or pandas. For JSON we use json.dump and json.load for files, and json.dumps and json.loads for strings."

1️⃣ open() — File Modes

```python
f = open("file.txt", "r")   # read — file must exist
f = open("file.txt", "w")   # write — creates new, deletes existing content
f = open("file.txt", "a")   # append — adds to end, keeps existing content
f = open("file.txt", "x")   # create — error if file already exists
```

2️⃣ read(), readline(), readlines()

```python
with open("file.txt", "r") as f:
    content = f.read()        # reads entire file as one string

with open("file.txt", "r") as f:
    line = f.readline()       # reads one line at a time

with open("file.txt", "r") as f:
    lines = f.readlines()     # reads all lines as a list

with open("file.txt", "r") as f:
    for line in f:            # best way — reads line by line
        print(line.strip())
```

3️⃣ write() and append()


```python
# write — creates file, overwrites if exists
with open("file.txt", "w") as f:
    f.write("Hello Rashid\n")
    f.write("ML Engineer\n")

# append — adds to end without deleting
with open("file.txt", "a") as f:
    f.write("New line added\n")
```

4️⃣ with statement

```python
# ❌ Old way — manual close, risky
f = open("file.txt", "r")
content = f.read()
f.close()           # if you forget this — memory leak

# ✅ Correct way — auto closes even if error occurs
with open("file.txt", "r") as f:
    content = f.read()
# file is automatically closed here
```

5️⃣ Exception Handling with Files


```python
try:
    with open("file.txt", "r") as f:
        content = f.read()
        print(content)

except FileNotFoundError:
    print("File does not exist")

except PermissionError:
    print("No permission to read this file")
```

6️⃣ CSV Files


```python
import csv

# Writing CSV
with open("students.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Marks", "Grade"])   # header
    writer.writerow(["Rashid", 92, "A"])
    writer.writerow(["Ali", 78, "B"])

# Reading CSV
with open("students.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)

# CSV with pandas — easiest way
import pandas as pd
df = pd.read_csv("students.csv")
print(df)
df.to_csv("output.csv", index=False)
```

7️⃣ JSON Files


```python
import json

# dumps — dict to JSON string (memory)
data = {"name": "Rashid", "age": 21}
json_string = json.dumps(data)
print(type(json_string))   # str

# loads — JSON string to dict (memory)
parsed = json.loads(json_string)
print(type(parsed))        # dict

# dump — dict to JSON file (file)
with open("data.json", "w") as f:
    json.dump(data, f, indent=4)

# load — JSON file to dict (file)
with open("data.json", "r") as f:
    loaded = json.load(f)
print(loaded["name"])      # Rashid
```

```python
# json.dumps — dict to string

import json

data = {"name": "Rashid", "age": 21}

json_string = json.dumps(data)
print(json_string)          # {"name": "Rashid", "age": 21}
print(type(json_string))    # <class 'str'>
```

```python
# json.loads — string to dict

import json

json_string = '{"name": "Rashid", "age": 21}'

data = json.loads(json_string)
print(data["name"])         # Rashid
print(type(data))           # <class 'dict'>
```

```python
# json.dump — dict to file

import json

data = {"name": "Rashid", "age": 21, "goal": "ML Engineer"}

with open("student.json", "w") as f:
    json.dump(data, f, indent=4)   # indent=4 makes file clean and readable
```

```python
# json.load — file to dict

import json

with open("student.json", "r") as f:
    data = json.load(f)

print(data["name"])    # Rashid
print(data["goal"])    # ML Engineer
```
