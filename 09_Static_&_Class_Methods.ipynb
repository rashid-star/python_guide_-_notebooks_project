"Instance methods take self and work on object data. Class methods take cls and work on class level data — they are used to change class variables or create alternative constructors. Static methods are utility functions that don't need access to class or object data at all. They are called directly on the class."

1. **Instance** method Need object data — self
2. **Class** methodNeed class data — cls, alternative constructors
3. **Static** methodUtility function — no class or object data needed text

```python
# Instance vs Class vs Static — comparison

class MyClass:
    class_var = "I am class variable"

    def instance_method(self):      # works on object data
        return self.class_var

    @classmethod
    def class_method(cls):          # works on class data
        return cls.class_var

    @staticmethod
    def static_method():            # works on nothing — just utility
        return "I am static"
```

```python
#### @classmethod — works on class, not object

class Student:
    school = "ABC School"           # class variable

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    @classmethod
    def change_school(cls, new_school):
        cls.school = new_school     # changes for ALL objects

    @classmethod
    def from_string(cls, data):     # alternative constructor
        name, marks = data.split(",")
        return cls(name, int(marks))

s1 = Student("Rashid", 92)
print(Student.school)               # ABC School

Student.change_school("XYZ School")
print(Student.school)               # XYZ School — changed for all

s2 = Student.from_string("Ali,85") # create object from string
print(s2.name, s2.marks)           # Ali 85
```

```python
### @staticmethod — utility function, no class or object needed

class MathUtils:
    @staticmethod
    def add(a, b):
        return a + b

    @staticmethod
    def is_even(n):
        return n % 2 == 0

    @staticmethod
    def celsius_to_fahrenheit(c):
        return (c * 9/5) + 32

# No object needed — call directly on class
print(MathUtils.add(5, 3))
print(MathUtils.is_even(4))
print(MathUtils.celsius_to_fahrenheit(100))
```

```python
# Create a class Temperature with:

# __init__ — takes celsius
# __str__ — returns "Temperature: {celsius}°C"
## @classmethod from_fahrenheit(cls, f) — creates object from fahrenheit value. Formula: c = (f - 32) * 5/9
## @classmethod from_kelvin(cls, k) — creates object from kelvin. Formula: c = k - 273.15
## @staticmethod is_freezing(celsius) — returns True if celsius <= 0
```

```python
class Temperature:
  def __init__(self, celsius):
    self.celsius = celsius

  def __str__(self):
    return f"Temperature: {self.celsius}°C"

  @classmethod
  def from_fahrenheit(cls, f):
    c = (f - 32) * 5/9
    return cls(c)

  @classmethod
  def from_kelvin(cls, k):
    c = k - 273.15
    return cls(c)

  @staticmethod
  def is_freezing(celsius):
    return celsius <= 0

t1 = Temperature(100)
t2 = Temperature.from_fahrenheit(212)
t3 = Temperature.from_kelvin(373.15)

print(t1)
print(t2)
print(t3)
print(Temperature.is_freezing(0))    # True
print(Temperature.is_freezing(25))   # False
```

```python
# Create a class Cart with:

# __init__ — takes owner and items as a list
# __str__ — returns "Cart of {owner} has {n} items"
# __len__ — returns number of items in cart
# __add__ — combines two carts and returns new cart with all items
# Method show_items() — prints all items
```

```python
class Cart:
  def __init__(self, owner, items):
    self.owner = owner
    self.items = items

  def __str__(self):
    return f"Cart of {self.owner} has {self.items}"

  def __len__(self):
    return len(self.items)

  def __add__(self, others):
    combines = self.items + others.items
    return Cart(self.owner, combines)

```

```python
# items is a list
c1 = Cart("Rashid", ["Books", "Pen", "Bag"])
c2 = Cart("Ali", ["Laptop", "Mouse"])

print(c1)           # calls __str__
print(len(c1))      # calls __len__
c3 = c1 + c2        # calls __add__
print(c3)
# c3.show_items()
```

```python
# Create a class Book with:

# __init__ — takes title, author, pages
# __str__ — returns "Book: {title} by {author}"
# __len__ — returns number of pages
# __eq__ — two books are equal if they have same title
# __add__ — adds two books and returns a new book with combined pages, same author, title = "{title1} & {title2}"
```

```python
class Book:
  def __init__(self, title, author, pages):
    self.title = title
    self.author = author
    self.pages = pages

  def __str__(self):
    return f"Book {self.title} by {self.author}"

  def __len__(self):
    return self.pages

  def __eq__(self, others):
    return self.title == others.title

  def __add__(self, other):
    new_title = self.title + " & " + other.title
    new_pages = self.pages + other.pages
    return Book(new_title, self.author, new_pages)

b1 = Book("Python Basics", "Rashid", 200)
b2 = Book("ML Guide", "Rashid", 350)
b3 = Book("Python Basics", "Ali", 200)

print(b1)           # __str__
print(len(b1))      # __len__
print(b1 == b2)     # False — different title
print(b1 == b3)     # True — same title
b4 = b1 + b2        # __add__
print(b4)
print(len(b4))
```
