```python
# 4️⃣ __name__ == "__main__"
'''
Every Python file has a __name__ variable.

When you run the file directly → __name__ == "__main__"
When you import the file → __name__ == "filename"
'''

# calculator.py

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

if __name__ == "__main__":
    # This runs ONLY when you run this file directly
    # NOT when someone imports it
    print(add(5, 3))
    print(subtract(10, 4))
```

```python
# Why this matters in real projects

# Without __name__ check — runs even when imported
print(add(5, 3))      # runs when imported — BAD

# With __name__ check — safe
if __name__ == "__main__":
    print(add(5, 3))  # only runs when executed directly — GOOD
```
