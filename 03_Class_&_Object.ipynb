```python
# Class = Blueprint
# Object = Real thing built from blueprint

class Car:
    def __init__(self, brand, color, speed):  # constructor
        self.brand = brand    # instance variable
        self.color = color
        self.speed = speed

    def describe(self):       # method
        print(f"{self.color} {self.brand} runs at {self.speed} km/h")

# Creating objects
car1 = Car("Toyota", "Red", 180)
car2 = Car("BMW", "Black", 250)

car1.describe()
car2.describe()
```

What is self?
self means this specific object. When you call car1.describe(), self = car1. When you call car2.describe(), self = car2.

```python
class Student:
    school = "ABC School"     # class variable — same for ALL objects

    def __init__(self, name, marks):
        self.name = name      # instance variable — different for each object
        self.marks = marks

s1 = Student("Rashid", 92)
s2 = Student("Ali", 78)

print(s1.school)   # ABC School
print(s2.school)   # ABC School — same for both
print(s1.name)     # Rashid
print(s2.name)     # Ali — different
```

```python
# Task 1: Create a class Phone with attributes brand, model, price. Add a method display() that prints all details. Create 2 objects.

class phone:
  def __init__(self, brand, model, price):
    self.brand = brand
    self.model = model
    self.price = price

  def display(self):
    print(f" {self.brand} {self.model} : {self.price}")

mob1 = phone("vivo", "X200", 80000)
mob2 = phone("oneplus", "15R", 40000)

mob1.display()
mob2.display()
```

```python
# Task 2: Create a class BankAccount with:

# Class variable bank_name = "ABC Bank"
# Instance variables owner and balance
# Method deposit(amount) that adds to balance
# Method withdraw(amount) that subtracts from balance — but if balance is less than amount print "Insufficient funds"
# Method show_balance() that prints current balance

class BankAccount:
  bank_name = "ABC Bank"

  def __init__(self, owner, balance):
    self.owner = owner
    self.balance = balance

  def deposit(self, amount):
      if amount > 0:
        self.balance += amount
        print(f"deposit of {amount} succesful. new balance for {self.owner} : {self.balance}")
      else:
        print("deposit amount must be possitive")

  def withdraw(self, amount):
    if amount <= 0:
      print("withdrawal amount must be possitive")
    elif self.balance < amount:
      print(f"Insufficient funds. current balance: {self.balance}, attemted withrawal : {amount}")
    else:
      self.balance -= amount
      print(f"Withdrawal of {amount} successful. New balance for {self.owner}: {self.balance}")

  def show_balance(self):
    print(f"Account balance for {self.owner} : {self.balance}")

my_account = BankAccount("Rashid", 2000)
my_account.show_balance()

my_account.deposit(1500)
my_account.withdraw(1000)
my_account.withdraw(10000)
my_account.show_balance()
```
