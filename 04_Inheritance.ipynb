"What is Inheritance?"

"Inheritance allows a child class to acquire properties and methods of a parent class. This promotes code reuse — we don't rewrite common logic. We use super() to call the parent class constructor. Python supports single, multilevel, and multiple inheritance."

Inheritance means — one class gets all properties and methods of another class.
Parent class = base. Child class = inherits everything + adds its own.


```python
# Parent class
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def eat(self):
        print(f"{self.name} is eating")

    def sleep(self):
        print(f"{self.name} is sleeping")

# Child class — inherits from Animal
class Dog(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age)  # calls parent __init__
        self.breed = breed           # extra attribute

    def bark(self):
        print(f"{self.name} is barking!")

d = Dog("Bruno", 3, "Labrador")
d.eat()    # inherited from Animal ✅
d.sleep()  # inherited from Animal ✅
d.bark()   # Dog's own method ✅
print(d.age)  # inherited from Animal ✅
```

What is super()?
super() calls the parent class method. Without it, parent __init__ never runs and name, age are never set.

```python
# Create a class Library with:

# Class variable library_name = "City Library"
# Instance variables book_title and available_copies
# Method borrow() — if copies > 0 reduce by 1 and print "Borrowed successfully", else print "No copies available"
# Method return_book() — increase copies by 1, print "Book returned"
# Method status() — print book title and available copies

class Library:
  library_name = "City Libaray"

  def __init__(self, book_title, available_copies):
    self.book_title = book_title
    self.available_copies = available_copies

  def borrow(self):
    if self.available_copies > 0:
        self.available_copies -= 1
        print(f"Borrowed successfully. Remaining copies: {self.available_copies}")  # ← this line missing
    else:
        print("No copies available")

  def return_book(self):
      self.available_copies += 1
      print("Book returned")

  def status(self):
    print(f"the {self.book_title} are {self.available_copies} pices")

libbooks = Library("The Fountain", 3)

libbooks.status()
libbooks.borrow()
libbooks.borrow()
libbooks.borrow()
libbooks.return_book()
libbooks.status()
```

```python
# Task 1: Create a parent class Vehicle with attributes brand and speed and a method move().
# Create a child class Car that adds attribute num_doors and a method honk().
# Create a child class Bike that adds attribute has_carrier and method wheelie(). Create one object of each and test all methods.

class Vahicle:
  def __init__(self, brand, speed):
    self.brand = brand
    self.speed = speed

  def move(self):
    print(f"{self.brand} is moving at {self.speed} km/h")

class Car(Vahicle):
  def __init__(self, brand, speed, num_doors):
    super().__init__(brand, speed)
    self.num_doors = num_doors

  def honk(self):
    print(f"{self.brand} is honking")

class Bike(Vahicle):
  def __init__(self, brand, speed, has_carrier):
    super().__init__(brand, speed)
    self.has_carrier = has_carrier

  def wheelie(self):
    print(f"{self.brand} is doing wheelie")

moter1 = Vahicle("BMW", 300)
car1 = Car("Toyota", 350, 4)
bike1 = Bike("Honda", 120, True)

moter1.move()
car1.move()
car1.honk()
bike1.move()
bike1.wheelie()
```

```python
class Animal:
    def breathe(self):
        print("Breathing")

class Dog(Animal):
    def bark(self):
        print("Barking")

class Puppy(Dog):       # inherits from Dog which inherits from Animal
    def cry(self):
        print("Crying")

p = Puppy()
p.breathe()  # from Animal ✅
p.bark()     # from Dog ✅
p.cry()      # own method ✅
```

```python
# Task 2: Create a class Person with name and age. Create a child class Employee that adds company and salary.
# Add a method get_info() in Employee that prints all 4 details — including name and age from parent.

class Person:

  def __init__(self, name, age):
    self.name = name
    self.age = age

class Employee(Person):
  def __init__(self,name, age, company, salary):
    super().__init__(name, age)
    self.company = company
    self.salary = salary

  def get_info(self):
    print(f"{self.name} is {self.age} and work in {self.company} and salary is {self.salary}")

me = Employee("Rashid", 21, "TCS", 30000)
me.get_info()
```
