"What are Dunder methods?"

"Dunder methods are special methods in Python with double underscores on both sides like str and len. They are called automatically by Python in specific situations. For example str is called when we print an object, len when we call len(), and add when we use the + operator. They allow us to define custom behavior for built-in operations."

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

# Without __str__
std = Student("Rashid", 92)
print(std)   # <__main__.Student object at 0x7f...> — ugly and useless
```

```python
# __str__ — what prints when you print an object

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __str__(self):
        return f"Student: {self.name}, Marks: {self.marks}"

std = Student("Rashid", 92)
print(std)   # Student: Rashid, Marks: 92 ✅
```

```python
# __len__ — what happens when you call len() on object

class Classroom:
    def __init__(self, students):
        self.students = students

    def __len__(self):
        return len(self.students)

c = Classroom(["Rashid", "Ali", "Sara", "Ahmed"])
print(len(c))   # 4 ✅
```

```python
# __add__ — what happens when you use + on objects

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2      # calls __add__ automatically
print(v3)          # Vector(6, 8) ✅
```

```python
# __eq__ — what happens when you use == on objects

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __eq__(self, other):
        return self.marks == other.marks

s1 = Student("Rashid", 92)
s2 = Student("Ali", 92)
s3 = Student("Sara", 78)

print(s1 == s2)   # True — same marks
print(s1 == s3)   # False — different marks
```

```python
# Create a class Cart with:

# __init__ — takes owner and items as a list
# __str__ — returns "Cart of {owner} has {n} items"
# __len__ — returns number of items in cart
# __add__ — combines two carts and returns new cart with all items
# Method show_items() — prints all items
```

```python
class Cart:
  def __init__(self, owner, items):
    self.owner = owner
    self.items = items

  def __str__(self):
    return f"Cart of {self.owner} has {self.items}"

  def __len__(self):
    return len(self.items)

  def __add__(self, others):
    combines = self.items + others.items
    return Cart(self.owner, combines)

```

```python
# items is a list
c1 = Cart("Rashid", ["Books", "Pen", "Bag"])
c2 = Cart("Ali", ["Laptop", "Mouse"])

print(c1)           # calls __str__
print(len(c1))      # calls __len__
c3 = c1 + c2        # calls __add__
print(c3)
# c3.show_items()
```

```python
# Create a class Book with:

# __init__ — takes title, author, pages
# __str__ — returns "Book: {title} by {author}"
# __len__ — returns number of pages
# __eq__ — two books are equal if they have same title
# __add__ — adds two books and returns a new book with combined pages, same author, title = "{title1} & {title2}"
```

```python
class Book:
  def __init__(self, title, author, pages):
    self.title = title
    self.author = author
    self.pages = pages

  def __str__(self):
    return f"Book {self.title} by {self.author}"

  def __len__(self):
    return self.pages

  def __eq__(self, others):
    return self.title == others.title

  def __add__(self, other):
    new_title = self.title + " & " + other.title
    new_pages = self.pages + other.pages
    return Book(new_title, self.author, new_pages)

b1 = Book("Python Basics", "Rashid", 200)
b2 = Book("ML Guide", "Rashid", 350)
b3 = Book("Python Basics", "Ali", 200)

print(b1)           # __str__
print(len(b1))      # __len__
print(b1 == b2)     # False — different title
print(b1 == b3)     # True — same title
b4 = b1 + b2        # __add__
print(b4)
print(len(b4))
```
