## Exception Handling


 Exception Handling
When your code breaks — Python throws an exception.
Without handling it, your entire program crashes.
Exception handling prevents crashes and gives clean error messages.

"Exception handling in Python is done using try, except, else, and finally blocks. The try block contains code that might fail. The except block handles specific errors. The else block runs only when no error occurs. The finally block always runs regardless of error. We can also raise exceptions manually using the raise keyword and create custom exceptions by inheriting from the Exception class."

```python
# Basic Structure

# try:
#     # code that might cause error
# except:
#     # what to do if error occurs
# else:
#     # runs only if NO error occurred
# finally:
#     # ALWAYS runs — error or no error
```

```python
# Common Exceptions You Must Know

int("hello")          # ValueError
10 / 0                # ZeroDivisionError
name = x              # NameError — x not defined
list = [1,2,3]
list[10]              # IndexError
dict = {"a": 1}
dict["b"]             # KeyError
int(None)             # TypeError
```

```python
# Catching Specific Exceptions

try:
    number = int(input("Enter a number: "))
    result = 10 / number
    print(f"Result: {result}")

except ValueError:
    print("That is not a valid number!")

except ZeroDivisionError:
    print("Cannot divide by zero!")

except Exception as e:
    print(f"Unexpected error: {e}")

else:
    print("No errors occurred!")

finally:
    print("Program finished.")
```

```python
# raise — Manually throw an exception

def set_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative")
    if age > 150:
        raise ValueError("Age is not realistic")
    return age

try:
    print(set_age(-5))
except ValueError as e:
    print(f"Error: {e}")
```

```python
# Custom Exceptions
# Create your own exception class

class InvalidMarksError(Exception):
    pass

def set_marks(marks):
    if marks < 0 or marks > 100:
        raise InvalidMarksError(f"Marks {marks} is out of range 0-100")
    return marks

try:
    set_marks(150)
except InvalidMarksError as e:
    print(f"Custom Error: {e}")
```

```python
# Task 2: get_student_marks(marks_list, index) — handles IndexError and ValueError

marks_list = [92, 83, 90, 79, 76]
def get_student_marks(marks_list, index):
    return marks_list[index]

try:
  print(get_student_marks(marks_list, 2))

except IndexError:
    print("Index out of range")

except TypeError:
  print("Index must be an integer")

try:
  print(get_student_marks(marks_list, 10)  )

except IndexError:
    print("Index out of range")

except TypeError:
  print("Index must be an integer")

try:
  print(get_student_marks(marks_list, "two"))

except IndexError:
    print("Index out of range")

except TypeError:
  print("Index must be an integer")
```

```python
# Task 3: Custom exception AgeError — raises if age below 5 or above 25

class AgeError(Exception):
    pass

def register_student(name, age):
    if age < 5 or age > 25:
        raise AgeError(f"Age {age} should between 5 to 25" )
        return age

try:
    register_student("Rashid", 21)
    register_student("Ali", 3)
    register_student("Sara", 30)
except AgeError as e:
    print(f"Error: {e}")
```
