"What is Encapsulation?"

"Encapsulation means restricting direct access to an object's data and methods. In Python we use single underscore for protected and double underscore for private variables. Private variables cannot be accessed directly from outside the class. We use getter and setter methods to read and update them safely with validation."

```python
# Encapsulation means — hiding internal data and protecting it from outside access.Real life
# example — ATM machine. You can deposit, withdraw, check balance. But you cannot directly access the internal cash system. It is hidden and protected.
```

```python
# Access 3 Levels in Python

class BankAccount:
    def __init__(self):
        self.name = "Rashid"        # public — anyone can access
        self._balance = 5000        # protected — only class and subclass
        self.__pin = 1234           # private — only this class
```

```python
acc = BankAccount()

print(acc.name)      # ✅ works
print(acc._balance)  # ⚠️ works but convention says don't access directly
print(acc.__pin)    # ❌ ERROR — private, cannot access directly
```

```python
# How to access private variables — Getters and Setters

class BankAccount:
    def __init__(self, name, balance, pin):
        self.name = name
        self._balance = balance
        self.__pin = pin

    # getter — read private value
    def get_pin(self):
        return self.__pin

    # setter — update private value with validation
    def set_pin(self, old_pin, new_pin):
        if old_pin == self.__pin:
            self.__pin = new_pin
            print("PIN updated successfully")
        else:
            print("Wrong old PIN")

    def show_balance(self):
        print(f"Balance: {self._balance}")

acc = BankAccount("Rashid", 5000, 1234)

print(acc.get_pin())           # ✅ 1234
acc.set_pin(1234, 9999)        # ✅ PIN updated
acc.set_pin(1234, 0000)        # ❌ Wrong old PIN
```

```python
acc2 = BankAccount("chaudhary", 10000, 9727)
print(acc2.get_pin())
acc2.set_pin(9727, 9721)
```

```python
# Create a class Student with:

# Public variable name
# Protected variable _grade
# Private variable __marks
# Getter get_marks() that returns marks
# Setter set_marks(new_marks) that updates marks — but only if new_marks is between 0 and 100, else print "Invalid marks"
# Method show_info() that prints all 3 variables

class Student:
  def __init__(self, name, grade, marks):
    self.name = name
    self._grade = grade
    self.__marks = marks

  def get_marks(self):
    return self.__marks

  def set_marks(self, new_marks):
    if 0 <= new_marks <= 100:
      self.__marks = new_marks
      print("Marks updated sucsessfully")
    else:
      print("Invalid marks")

  def show_info(self):
    print(f"Marks : {self.__marks}")

std = Student("Rashid", "A", 92)

std.get_marks()
std.set_marks(95)
std.show_info()

# Test all of these
print(std.name)        # public — direct access ✅
# print(std.__marks)     # private — should give ERROR ❌
print(std.get_marks()) # getter ✅
std.set_marks(95)      # valid ✅
std.set_marks(150)     # invalid ✅
std.show_info()        # all info ✅
```

```python
# Create a class Hospital with:

# Public variable hospital_name
# Protected variable _department
# Private variable __patient_records
# Getter get_records() — returns patient records
# Setter set_records(new_record) — adds new record only if it is a non-empty string, else print "Invalid record"
# Method show_info() — prints all 3 variables

# Test it:

# Access hospital_name directly ✅
# Try accessing __patient_records directly ❌
# Use getter ✅
# Use setter with valid record ✅
# Use setter with empty string "" ✅
# Call show_info() ✅
```

```python
class Hospital:
  def __init__(self, hospital_name, department, patient_records):
     self.hospital_name = hospital_name
     self._department = department
     self.__patient_records = patient_records

  def get_records(self):
    return self.__patient_records

  def set_records(self, new_record):
    if new_record != "":
      self.__patient_records = new_record
      print("record updated sucsessfully")

    else:
      print("Invalid record")

  def show_info(self):
    print(f"record : {self.__patient_records}")

pt = Hospital("JJ Hospital", "Ortho", "Rashid - back_pain")

print(pt.get_records())
pt.set_records("Hammad - fever")
print(pt.get_records())
```
