## Generators & yield


"A generator uses yield instead of return. It pauses execution at yield and resumes from the same point on the next call. This saves memory because values are generated one at a time instead of all at once. In ML this is very useful for loading large datasets in batches."

Normal function returns everything at once.
Generator pauses at yield and resumes from same point next call.

```python
# Normal function — creates entire list in memory
def normal_squares(n):
    result = []
    for i in range(n):
        result.append(i ** 2)
    return result


# Generator — generates one value at a time
def gen_squares(n):
    for i in range(n):
        yield i ** 2       # pauses here, resumes next call

# Usage
gen = gen_squares(5)
print(next(gen))   # 0
print(next(gen))   # 1
print(next(gen))   # 4

# Or loop through it
for val in gen_squares(5):
    print(val)
```

```python
# Why generators matter in ML
# When dataset has millions of rows — you can't load all in memory. Generator loads one batch at a time.

def data_loader(data, batch_size):
    for i in range(0, len(data), batch_size):
        yield data[i:i + batch_size]   # yields one batch at a time

data = list(range(100))

for batch in data_loader(data, 10):
    print(batch)   # processes 10 items at a time
```

```python
# Task 1: Write a generator even_numbers(n) that yields even numbers from 0 to n. Loop through and print all values.

def even_numbers(n):
  for i in range(0, n):
    if i % 2 == 0:
      yield i

for val in even_numbers(20):
  print(val)
```

```python
# Task 2: Write a generator countdown(n) that counts down from n to 0 using yield. Test with countdown(10).

def countdown(n):
  for i in range(n, -1, -1):
    yield i

for val in countdown(22):
  print(val)
```
